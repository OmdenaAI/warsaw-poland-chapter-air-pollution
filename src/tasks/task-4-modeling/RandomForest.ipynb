```python
df_aq_weather_static_merged = pd.read_csv("Combined_air_pollutants_data_with_caqi.csv")
df_aq_weather_static_merged.head()
```

```python
df = df_aq_weather_static_merged.drop(['index', 'LAT', 'LON','voivodship','powiat_voivod', 'NO2_24H_AVG_POLLUTION', 'O3_24H_AVG_POLLUTION', 'PM10_24H_AVG_POLLUTION', 'PM25_24H_AVG_POLLUTION',
        'NO2_sub_idx', 'O3_sub_idx', 'PM10_sub_idx', 'PM25_sub_idx','CAQI_idx'], axis=1)
df.head()
```

```python
df['DATE']=df['DATE'].astype('datetime64')
```

```python
df['date_offset'] = (df.DATE.dt.month*100 + df.DATE.dt.day - 320)%1300

df['season'] = pd.cut(df['date_offset'], [0, 300, 602, 900, 1300], 
                      labels=['spring', 'summer', 'autumn', 'winter'])
df = df.drop(['date_offset', 'DATE'], axis=1)
```

```python
data = df.copy()
target = data.pop('CAQI_level')
data.head()
data = pd.get_dummies(data, columns=['season' ], drop_first=True).astype('float')
data.head()
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target)

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(min_samples_leaf = 3, n_estimators=600, random_state=42)

classifier.fit(data, target)
```

```python
classifier.score(data, target)
```

```python
from sklearn.metrics import accuracy_score

y_pred = classifier.predict(X_test)

accuracy_score(y_test, y_pred)
```
