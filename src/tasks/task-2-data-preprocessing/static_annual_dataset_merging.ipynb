```python
import glob
import os
import pandas as pd
```

```python
readdict = {}
readdict[('County Code', 'Unnamed: 0_level_1', 'Unnamed: 0_level_2')] = str
readdict[('Voivodeship Code', 'Unnamed: 0_level_1', 'Unnamed: 0_level_2')] = str
readdict[('County Code','Unnamed: 0_level_1','Unnamed: 0_level_2','Unnamed: 0_level_3')] = str
readdict[('Voivodeship Code','Unnamed: 0_level_1','Unnamed: 0_level_2','Unnamed: 0_level_3')] = str
readdict[('County Code','Unnamed: 0_level_1','Unnamed: 0_level_2','Unnamed: 0_level_3','Unnamed: 0_level_4')] = str
readdict[('Voivodeship Code','Unnamed: 0_level_1','Unnamed: 0_level_2','Unnamed: 0_level_3','Unnamed: 0_level_4')] = str
```

```python
key = [('County Code', 'Unnamed: 0_level_1','Unnamed: 0_level_2')]
```

### Animal stock + area by land use

```python
df = pd.read_excel("../../data/data_processed/static_annual_data/animal_stock_by_voivodship_dist.xlsx",header=[0,1,2], \
                   dtype = readdict)
#Remove first column, rename total and remove first row
df.drop(columns=df.columns[0],inplace=True)
df.rename(columns={'total':"total animal stock"},inplace=True)
df.drop(index=0,inplace=True)
```

```python
#Read second file
df2 = pd.read_excel("../../data/data_processed/static_annual_data/area_by_land_use_by_voivodship_dist.xlsx", header=[0,1,2], \
                   dtype = readdict)
df2.rename(columns={'total in ha':"total area", "forests": "forests area", "wasteland": "wasteland area", },inplace=True)
df2.drop(columns=df2.columns[0],inplace=True)
df2.drop(index=0,inplace=True)
```

```python
def drop_duplicates(merged_df):
    cols_to_drop = []
    for col in merged_df.columns:
        if col[0].endswith('_right'):
            cols_to_drop.append(col)
    merged_df = merged_df.drop(columns=cols_to_drop)
    return merged_df
```

```python
m1 = pd.merge(df,df2,on=key,suffixes=[None,"_right"],validate="1:1")
m1 = drop_duplicates(m1)
```

### + area_by_powiat

```python
df2 = pd.read_excel("../../data/data_processed/static_annual_data/area_by_powiat.xlsx", header=[0,1,2], \
                   dtype = readdict)
df2.rename(columns={'total in square km':"total area in square km"},inplace=True)
df2.drop(columns=df2.columns[0],inplace=True)
df2.drop(index=0,inplace=True)
```

```python
m1 = pd.merge(m1,df2)
m1 = drop_duplicates(m1)
```

### + emission_of_particulates_by_powiat

```python
df2 = pd.read_excel("../../data/data_processed/static_annual_data/emission_of_particulates_by_powiat.xlsx", header=[0,1,2], \
                   dtype = readdict)
df2.rename(columns={'total':"emission_particulate total", 'total per 1 km2 of area': 'emission_particulate total per 1 km2 of area',\
                    'fugitive':'emission_particulate fugitive', 'fuel combustion':'emission_particulate fuel combustion',\
                    'cement/lime and refractory materials':'emission_particulate cement/lime and refractory materials',\
                    'silicon': 'emission_particulate silicon', 'chemical fertilizers':'emission_particulate chemical fertilizers',\
                    'carbon and graphite, soot':'emission_particulate carbon and graphite, soot' },inplace=True)
df2.drop(columns=df2.columns[0],inplace=True)
df2.drop(index=0,inplace=True)
```

```python
m1 = pd.merge(m1,df2)
m1 = drop_duplicates(m1)
```
