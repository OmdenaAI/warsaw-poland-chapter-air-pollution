```python
import pandas as pd
```

```python
readdict = {}
readdict[('County Code', 'Unnamed: 0_level_1', 'Unnamed: 0_level_2')] = str
readdict[('Voivodeship Code', 'Unnamed: 0_level_1', 'Unnamed: 0_level_2')] = str
readdict[('County Code','Unnamed: 0_level_1','Unnamed: 0_level_2','Unnamed: 0_level_3')] = str
readdict[('Voivodeship Code','Unnamed: 0_level_1','Unnamed: 0_level_2','Unnamed: 0_level_3')] = str
readdict[('County Code','Unnamed: 0_level_1','Unnamed: 0_level_2','Unnamed: 0_level_3','Unnamed: 0_level_4')] = str
readdict[('Voivodeship Code','Unnamed: 0_level_1','Unnamed: 0_level_2','Unnamed: 0_level_3','Unnamed: 0_level_4')] = str
```

# Mering files with 3 level MultiIndex

```python
files3 = ['animal_stock_by_voivodship_dist.xlsx','area_by_land_use_by_voivodship_dist.xlsx','area_by_powiat.xlsx','emission_of_particulates_by_powiat.xlsx','emission_of_pollutant_gases_by_powiat.xlsx',\
          'forest_area_by_powiat.xlsx', 'plants_of_significant_nuisance_to_air_quality_by_voivodship_dist.xlsx', 'population_density_by_powiat.xlsx', 'production_of_electricity_by_sources_by_voivodship_dist.xlsx',\
            'share_of_green_areas_by_powiat.xlsx','share_of_urbanized_area_by_voivodship_dist.xlsx','total_vehicles_by_type_by_powiat.xlsx']

prefix3 = ['[animal stock]', '[area by land]', '[area]', '[emission of particulates]','[emission of pollutant gases]', '[forest area]','[plants of significant nuisance]',\
           '[population density]', '[production of electricity]', '[share of green areas]', '[share of urbanized areas]', '[total vehicles]']

key3 = [('County Code', 'Unnamed: 0_level_1','Unnamed: 0_level_2')]
```

```python
def rename_columns(df, prefix):
    cols = {}
    for col in df.columns[4:]:
        if col[0] not in cols:
            cols[col[0]] = ' '.join([prefix,col[0]])
    return cols

def drop_duplicates(merged_df):
    cols_to_drop = []
    for col in merged_df.columns:
        if col[0].endswith('_right'):
            cols_to_drop.append(col)
    merged_df = merged_df.drop(columns=cols_to_drop)
    return merged_df
```

```python
df = pd.read_excel("../../data/data_processed/static_annual_data/"+files3[0],header=[0,1,2], \
                   dtype = readdict)
df.head()
```

```python
#Remove first column, add prefix and remove first row
df.drop(columns=df.columns[0],inplace=True)
df.rename(columns=rename_columns(df,prefix3[0]),inplace=True)
df.drop(index=0,inplace=True)
```

```python
#add isUrban feature
df['isUrban','Unnamed: 0_level_1','Unnamed: 0_level_2'] = \
    [1 if x[7].isupper() else 0 for x in df['County Name', 'Unnamed: 1_level_1', 'Unnamed: 1_level_2']]
```

```python
m3df= df
for i,fname in enumerate(files3):
    if i==0:
        continue
    df2 = pd.read_excel("../../data/data_processed/static_annual_data/"+fname, header=[0,1,2], \
                   dtype = readdict)
    df2.drop(columns=df2.columns[0],inplace=True)
    df2.rename(columns=rename_columns(df2,prefix3[i]),inplace=True)
    df2.drop(index=0,inplace=True)
    #merge
    m3df = pd.merge(m3df,df2,on=key3,suffixes=[None,"_right"],validate="1:1")
    m3df = drop_duplicates(m3df)
```

# Mering files with 4 level MultiIndex

```python
files4 = ['air_pollution_reduction_systems_in_plants_by_voivodship_dist.xlsx', 'forest_fires_by_voivodship_dist.xlsx', 'vehicles_by_type_and_fuel_by powiat.xlsx']

prefix4 = ['[air pollution reduction systems]', '[forest fires]', '[vehicles by type and fuel]']

key4 = [('County Code', 'Unnamed: 0_level_1','Unnamed: 0_level_2','Unnamed: 0_level_3')]
```

```python
df = pd.read_excel("../../data/data_processed/static_annual_data/"+files4[0],header=[0,1,2,3], \
                   dtype = readdict)
df.head()
```

```python
#Remove first column, add prefix and remove first row
df.drop(columns=df.columns[0],inplace=True)
df.rename(columns=rename_columns(df,prefix4[0]),inplace=True)
df.drop(index=0,inplace=True)
```

```python
m4df= df
for i,fname in enumerate(files4):
    if i==0:
        continue
    df2 = pd.read_excel("../../data/data_processed/static_annual_data/"+fname, header=[0,1,2,3], \
                   dtype = readdict)
    df2.drop(columns=df2.columns[0],inplace=True)
    df2.rename(columns=rename_columns(df2,prefix4[i]),inplace=True)
    df2.drop(index=0,inplace=True)
    #merge
    m4df = pd.merge(m4df,df2,on=key4,suffixes=[None,"_right"],validate="1:1")
    m4df = drop_duplicates(m4df)
```

# Mering files with 5 level MultiIndex

```python
files5 = ['crop_production_by_voivodship_dist.xlsx']

prefix5= ['[crop production]']

key5 = [('County Code', 'Unnamed: 0_level_1','Unnamed: 0_level_2','Unnamed: 0_level_3','Unnamed: 0_level_4')]
```

```python
df = pd.read_excel("../../data/data_processed/static_annual_data/"+files5[0],header=[0,1,2,3,4], \
                   dtype = readdict)
df.head()
```

```python
#Remove first column, add prefix and remove first row
df.drop(columns=df.columns[0],inplace=True)
df.rename(columns=rename_columns(df,prefix5[0]),inplace=True)
df.drop(index=0,inplace=True)
m5df = df
```

# Expand all MultiIndex to 5 levels

Convention: 
* first level contains the name with the prefix,
* last level contains unit
* penultimate level contains the year
* the rest is filled with blank levels 

```python
m3df_newmi = []
for i, t in enumerate(m3df.columns):
    if t[0] in ['County Code', 'Voivodeship Code','isUrban'] :
        temp = list(t) + ['Unnamed: 0_level_3','Unnamed: 0_level_4']
        m3df_newmi.append(tuple(temp))
    elif t[0] in ['County Name', 'Voivodeship']:
        temp = list(t) + ['Unnamed: 1_level_3','Unnamed: 1_level_4']
        m3df_newmi.append(tuple(temp))
    else:
        temp = list(t)
        assert len(temp) == 3
        temp_list = [temp[0], 'Unnamed: 0_level_1', 'Unnamed: 0_level_2', temp[1], temp[2]]
        m3df_newmi.append(tuple(temp_list))

ind = m3df_newmi.index(('isUrban', 'Unnamed: 0_level_1', 'Unnamed: 0_level_2', 'Unnamed: 0_level_3', 'Unnamed: 0_level_4'))
m3df_newmi = pd.MultiIndex.from_tuples(m3df_newmi)
```

```python
m4df_newmi = []
for i, t in enumerate(m4df.columns):
    if i < 4:
        if t[0] in ['County Code', 'Voivodeship Code']:
            temp = list(t) + ['Unnamed: 0_level_4']
        elif t[0] in ['County Name', 'Voivodeship']:
            temp = list(t) + ['Unnamed: 1_level_4']
        m4df_newmi.append(tuple(temp))
    else:
        temp = list(t)
        assert len(temp) == 4
        temp_list = [temp[0], 'Unnamed: 0_level_1', temp[1], temp[2], temp[3]]
        m4df_newmi.append(tuple(temp_list))
m4df_newmi = pd.MultiIndex.from_tuples(m4df_newmi)
```

```python
m3df.columns = m3df_newmi
m4df.columns = m4df_newmi
```

# Merge all data

```python
static_final = pd.merge(m3df,m4df,on=key5,suffixes=[None,"_right"],validate="1:1")
static_final = drop_duplicates(static_final)
static_final = pd.merge(static_final,m5df,on=key5,suffixes=[None,"_right"],validate="1:1")
static_final = drop_duplicates(static_final)
```

```python
#reorder columns
isUrbanColName = static_final.columns[ind]
isUrbanValues = static_final.iloc[:,ind]
static_final.drop(columns = [isUrbanColName],inplace=True)
static_final.insert(loc=4, column=isUrbanColName, value=isUrbanValues)
```

```python
#save the data
static_final.to_excel(f'../../data/data_processed/static_annual_data/combined_static_annual_data.xlsx')
```
